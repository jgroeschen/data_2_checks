```python
import pandas as pd
import requests
```

# Pull in data from an API

```python
url = 'https://datausa.io/api/data?drilldowns=State&measures=Population'
response = requests.get(url).json().get('data')
df = pd.DataFrame(response)
df
```

# Find and print TWO descriptive statistics about your data

```python
pop_mean = int(df['Population'].mean())
total2020 = df.loc[df['Year'] == '2020']['Population'].sum()
print(f'Mean state population (2013-2020): {pop_mean: ,}', f'US population total (2020): {total2020: ,}', sep='\n')
```

# Write a query in Pandas to select a particular set of your data

```python
df.loc[df['State'] == 'Kentucky'][['State', 'Year', 'Population']]
```
